# Preprocessing

#### Removing Null Rows

```python
df.isna().sum()
```

```python
df.dropna(inplace= True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.head()
```

#### Handling Null Value

```python
df["rating"].replace("--", np.nan, inplace= True)
```

```python
# Fill nan with -999, will help the model.
# If we would have used nan, model would have ignored the entire row.
# Restaurant with no ratig is very criticall, because we human usually avoid those places.

df["rating"].fillna(-999.0, inplace= True)
```

```python
df["rating"] = df["rating"].astype(float)
```

```python
df.info()
```

```python
df
```

### Handling categorical count of ratings

### Handling Special Symbol in Cost

```python
df["cost"] = df["cost"].str.replace("₹", "").astype("int")
```

### What is the rating distribution?

```python
ratings_df = 100 * df.rating.value_counts().sort_index()/len(df)
ratings_df = ratings_df.reset_index()
ratings_df.columns = ['rating', 'percentage']
ratings_df['rating'] = ratings_df['rating'].astype(str)
ratings_df.head()
```

```python
ratings_df.plot(x= "rating", y= "percentage", alpha= 0.5);
```

```python
# Removing the -999.0 rating from the plot

ratings_df = ratings_df.tail(-1)
ratings_df.plot(x= "rating", y= "percentage", alpha= 0.5);
```
