# **Data Cleaning**

```python
df.head()
```

```python
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".", "").astype(float)
```

```python
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".", "").str.replace(",", ".").astype(float)
```

```python
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".", "").astype(float)
```

```python
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].str.replace(".", "").str.replace(",", ".").astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
# 1 missing value in Acreage (Hectare) and 1 missing value in Yield (Kg / Hectare).

df.isna().sum()
```

```python
# Both missing value is in single row.

df[df["Acreage (Hectare)"].isnull()]
```

```python
df = df.dropna().reset_index()
```

```python
df.head()
```

```python
df.drop(columns= ["index", "Unnamed: 0"], inplace= True)
```
